```python
np.random.seed(42)
readings = np.random.normal(loc=50, scale = 5, size=(360,4))

scaled_readings = (readings - readings.min()) / (readings.max() - readings.min())

centered_readings = readings - readings.mean(axis=0)

print(readings.shape)
print(scaled_readings.shape)
print(centered_readings.shape)
```

Task 3

```python
import numpy as np

# Re-initializing with the same seed and parameters
np.random.seed(42)
readings = np.random.normal(loc=50, scale=5, size=(360, 4))

# 1. Build a boolean mask for outliers (> 80 or < 20)
outlier_mask = (readings > 80) | (readings < 20)

# 2. Compute outlier statistics
outlier_count = np.sum(outlier_mask)
outlier_percentage = (outlier_count / readings.size) * 100

# 3. Create cleaned_readings array with NaNs
cleaned_readings = readings.copy()
cleaned_readings[outlier_mask] = np.nan

# 4. Verification
nan_count = np.count_nonzero(np.isnan(cleaned_readings))

print(f"Outlier Count: {outlier_count}")
print(f"Outlier Percentage: {outlier_percentage:.2f}%")
print(f"Number of NaNs in cleaned array: {nan_count}")
print(f"Verification Successful: {outlier_count == nan_count}")
```

task 4

```python
sensor_means = readings.mean(axis=0)
sensor_std = readings.std(axis=0)

time_means = readings.mean(axis=1)

print(sensor_means.shape)
print(sensor_std.shape)
print(time_means.shape)
print(f"Average reading for Sensor 1: {sensor_means[0]:.2f}")
print(f"Average reading at Minute 1:  {time_means[0]:.2f}")
```

Task 5

```python
report = {
    'total_readings': readings.size,
    'outlier_count' : outlier_count,
    'outlier_percentage': outlier_percentage,
    'sensor_means' : sensor_means,
    'sensor_stds' : sensor_std
}
print(report)
#Validation check
print( report['total_readings']==readings.size)

string_report = str(report)
print(string_report)
```

```python
import numpy as np 
rng = np.random.default_rng(seed=0)
sim = rng.normal(0,1,size=(1000,6))
print(sim.shape)
print(sim.dtype)
print(sim[:3])
```

Task 7

```python
scenario_shift = np.array([-0.3,-0.1,0.0,0.1,0.2,0.4])
adjusted_sim = scenario_shift + sim
print(adjusted_sim)
print(adjusted_sim.shape==sim.shape)
```

Task 8

```python
scenario_means = np.mean(adjusted_sim, axis=0)

scenario_ranks = np.argsort(scenario_means)
partitioned_means = np.partition(scenario_means, -2)
top_two_partitioned = partitioned_means[-2:]

full_sort_top_two = np.sort(scenario_means)[-2:]
print(np.array_equal(np.sort(top_two_partitioned), full_sort_top_two))
```
